## Sumário Executivo
### Análise de Dados SAF-T Padaria.
Este projeto tem como objetivo analisar um conjunto de dados de vendas e impostos para identificar padrões, tendências e insights que possam ajudar a empresa a otimizar suas operações e estratégias de vendas. O dataset inclui informações sobre transações de vendas, códigos de impostos, produtos e outras variáveis relevantes.

**Aluno:** Karen Pessoa

**Data:** 19/02/2026


1. Período Analisado: 02 de Janeiro de 2024 a 08 de Dezembro de 2024.
Total Faturado: € 135345.04, sendo que deste total € 13074.89 são de impostos e taxas.
Ticket Médio Global: € 2.98.
Volume de Transações: 43127 vendas processadas.
Produtos vendidos: 428 itens diferentes comercializados.

2. Processo de Limpeza e Preparação
Estado Inicial: Os dados apresentavam descrições duplicadas, inconsistências de capitalização e formatos de data não padronizados.
Passos de Limpeza:
* Conversão de MovementStartTime para objeto datetime.
* Normalização de descrições para minúsculas para evitar contagem duplicada.
* Tratamento de outliers e valores nulos em GrossValue.
Estrutura Final: Criação de colunas derivadas como Dia_Semana, Categoria, Mês e Hora.

3. Análise de Vendas
3.1 Performance Temporal
Mês de Pico: Setembro com maior faturação, possivelmente impulsionado pelo período de férias / verão.
Mês de Baixa: Não podemos afirmar que Dezembro foi o mês que apresentou o menor volume, pois foram analisados apenas 8 dias. Neste caso, Janeiro foi o mês que apresentou menor volume, que de fato coincide com um período pós festas e de menor fluxo local.
Tendência: Oscila ao longo do ano analisado, sendo crescente no meio do ano e descresce ao final do ano.

3.2 Análise de Produtos
Top 5 Faturação: Café, pão pequeno caseiro, lucky strike, bolos para consumo interno e fofa.
Top 5 Quantidade: Café, fofa, bebés, pão pequeno caseiro, bola caseira.
Análise ABC: Aproximadamente 0.08% dos produtos (Classe A) geram 80% da receita, evidenciando uma dependência de itens "âncora" como Café e Pão.
Categorias: Produtos de mercearia formam a categoria mais rentável, enquanto produtos de padaria geram o maior volume de visitas.

3.3 Padrões de Comportamento
Dia Preferido: Domingo apresenta a maior faturação total.
Ticket Médio: Observou-se que o valor gasto por cliente aumenta aos fins de semana, indicando consumo familiar vs. consumo individual de rotina.

4. Insights e Recomendações
Insights Descobertos
 1. A análise evidenciou que uma pequena parcela dos produtos como sendo responsável pela maior parte da receita total, caracterizando um efeito semelhante ao princípio de Pareto. Isso indica alta dependência de produtos-chave para sustentação do faturamento.
***Implicação:*** vulnerabilidade do negócio a variações de procura, preço ou ruptura de estoque desses itens.
2. Em contrapartida com a alta faturação para uma parcela pequena de produtos, há parte do portfólio que apresenta baixa frequência de venda e contribuição contribui muito pouco para a receita, mas ainda consome espaço produtivo e logístico.
***Implicação:*** possível ineficiência na alocação de recursos.
3. Observou-se variação sistemática nas vendas ao longo do tempo, com picos concentrados em determinados períodos (dias da semana / meses específicos). A saber, sábados e domingos como os melhores dias de faturação da semana, e os meses de verão / férias como o melhor período de faturação do ano. 
***Implicação:*** existe oportunidade de planejamento mais eficiente de produção e estoque com base em padrões previsíveis.
4. O ticket médio apresenta estabilidade ao longo do período, sugerindo comportamento de consumo previsível. Entretanto, a receita total é fortemente influenciada pela frequência de compras. 
***Implicação:*** estratégias de retenção podem gerar maior impacto do que aumento pontual de preços.

Recomendações de Negócio
1. Garantir disponibilidade contínua dos produtos de maior receita, com controle rigoroso de estoque e possível otimização de precificação.
2. Avaliar descontinuação, reformulação ou reposicionamento dos produtos com baixa saída, reduzindo complexidade operacional.
3. Ajustar produção, escala de equipe e compras de insumos com base nos padrões temporais identificados, visando redução de desperdício e aumento de eficiência.
4. Cross-selling: Criar menus combinando produtos da Classe A (Café) com itens de Pastelaria de margem maior.
5. Promoções: Aplicar descontos em dias de baixo movimento para nivelar a faturação semanal.

5. Limitações da Análise
Dados em Falta: Falta de identificação única de cliente (Fidelização) para análise de recorrência.
Custos: Sem o preço de custo (COGS), a análise foca apenas em faturação bruta e não em lucro líquido.
Qualidade: Algumas descrições genéricas no sistema dificultaram a categorização 100% precisa de certos itens.
